### Heroes Of Pymoli Data Analysis

-Overall, the 20-24 year age group contributes the most to total revenue at 64.16%. This is also the largest demographic at 44.79%, yeilding a per capita spending of 41.64% above the average ($3.05/person). This means that 20-24 year olds are a very important demographic.

-The highest spending group per capita was the 35-39 year olds, spending on average 56.07% more than the average ($3.05/person). They are small portion of the player base (5.38% of players) and do not generate a large portion of revenue (6.21% of Total Revenue). This means that drawing in more players in this age range could be highly profitable.

-Of the 5 users that spent the most money, 4 of them were in the 20-24 age group, though of their purchases only two items from the Most Profitable Items and Most Popular Items lists appeared, once each from different users (Oathbreaker, Last Hope of the Breaking Storm and Nirvana). This means that the people that spend the most do not necessarily purchase the most popular or expensive items.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os

# File to Load (Remember to Change These)
file_to_load = os.path.join('Resources', 'purchase_data.csv')

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
# New DataFrame that stores the total number of unique players
total_player = pd.DataFrame([{'Total Players':len(purchase_df['SN'].unique())}])
total_player
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# New DataFrame that stores the Purchasing Analysis(Total) requirements
purchasing_analysis_total = pd.DataFrame([{
    'Number of Unique Items':len(purchase_df['Item ID'].unique()),
    'Average Price':purchase_df['Price'].mean(),
    'Number of Purchases':len(purchase_df['Purchase ID']),
    'Total Revenue':purchase_df['Price'].sum()
    }])

# Format the DataFrame
purchasing_analysis_total['Average Price'] = purchasing_analysis_total['Average Price'].map("${:.2f}".format)
purchasing_analysis_total['Total Revenue'] = purchasing_analysis_total['Total Revenue'].map("${:,.2f}".format)

# Show summary
purchasing_analysis_total
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# New DataFrame to store gender divided data
male_df = purchase_df.loc[purchase_df['Gender'] == 'Male',:]
female_df = purchase_df.loc[purchase_df['Gender'] == 'Female',:]
other_df = purchase_df.loc[purchase_df['Gender'] == 'Other / Non-Disclosed',:]
```

```python
# Variables to hold total number of unique SNs by gender
numMales = len(male_df['SN'].unique())
numFemales = len(female_df['SN'].unique())
numOther = len(other_df['SN'].unique())

# Summary DataFrame storing Gender Demographics requirements
gender_demo = pd.DataFrame({
    'Total Count':[numMales, numFemales, numOther],
    'Percentage of Players':[
        (numMales/total_player.iloc[0,0]*100),
        (numFemales/total_player.iloc[0,0]*100),
        (numOther/total_player.iloc[0,0]*100)]},
        index = ['Male', 'Female', 'Other/Non-Disclosed'])

# Format DataFrame
gender_demo['Percentage of Players'] = gender_demo['Percentage of Players'].map("{:.2f}%".format)
gender_demo.index.name = 'Gender'

# Show summary
gender_demo
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# New DataFrame to store requirements for Purchasing Analysis(Gender)
pur_analysis_gender = pd.DataFrame({
    'Purchase Count':[
        len(male_df),
        len(female_df),
        len(other_df)],
    'Total Purchase Value':[
        male_df['Price'].sum(),
        female_df['Price'].sum(),
        other_df['Price'].sum()],
    'Average Purchase Price':[
        (male_df['Price'].sum() / len(male_df)),
        (female_df['Price'].sum() / len(female_df)),
        (other_df['Price'].sum() / len(other_df))],
    'Avg Total Purchase per Person':[
        (male_df['Price'].sum() / numMales),
        (female_df['Price'].sum() / numFemales),
        (other_df['Price'].sum() / numOther)]},
    index = ['Male', 'Female', 'Other/Non-Disclosed'])

# Format DataFrame
pur_analysis_gender['Average Purchase Price'] = pur_analysis_gender['Average Purchase Price'].map("${:.2f}".format)
pur_analysis_gender['Total Purchase Value'] = pur_analysis_gender['Total Purchase Value'].map("${:,.2f}".format)
pur_analysis_gender['Avg Total Purchase per Person'] = pur_analysis_gender['Avg Total Purchase per Person'].map("${:.2f}".format)
pur_analysis_gender.index.name = 'Gender'

# Show summary
pur_analysis_gender
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Define bins and group names to divide data with
bins = [0, 9, 14, 19, 24, 29, 34, 39, 150]
group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

# Make a DataFrame with no duplicate SNs
binned_df = purchase_df.drop_duplicates(subset='SN')

# Divide data into bins
binned_df['Binned Ages'] = pd.cut(binned_df['Age'], bins, labels=group_names)
bin_counts = binned_df['Binned Ages'].value_counts()

# Make and populate summary DataFrame
age_demo_df = pd.DataFrame({
    'Total Count':bin_counts,
    'Percentage of Players':bin_counts/total_player.iloc[0,0]*100},
    index=group_names)

# Format DataFrame
age_demo_df['Percentage of Players'] = age_demo_df['Percentage of Players'].map("{:.2f}%".format)
age_demo_df.index.name = 'Age Group'

# Show summary
age_demo_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Bin original DataFrame
all_ages_df = purchase_df
all_ages_df['Binned Ages'] = pd.cut(all_ages_df['Age'], bins, labels=group_names)

# Variables to store info for easy use later
pcAll = all_ages_df.groupby('Binned Ages')['Purchase ID'].count()
tpvAll = all_ages_df.groupby('Binned Ages')['Price'].sum()

# Make and populate summary DataFrame
purchase_analysis_age_df = pd.DataFrame({
    'Purchase Count': pcAll,
    'Total Purchase Value': tpvAll,
    'Average Purchase Price': tpvAll / pcAll,
    'Avg Total Purchase per Person': tpvAll / binned_df.groupby('Binned Ages')['SN'].count()
})

# Format DataFrame
purchase_analysis_age_df['Average Purchase Price'] = purchase_analysis_age_df['Average Purchase Price'].map("${:.2f}".format)
purchase_analysis_age_df['Total Purchase Value'] = purchase_analysis_age_df['Total Purchase Value'].map("${:,.2f}".format)
purchase_analysis_age_df['Avg Total Purchase per Person'] = purchase_analysis_age_df['Avg Total Purchase per Person'].map("${:.2f}".format)
purchase_analysis_age_df.index.name = 'Age Group'

# Show Summary
purchase_analysis_age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Group by SN
user_group = purchase_df.groupby('SN')

# Variables to store values for easy access later
user_purchases = user_group['Purchase ID'].count()
user_total = user_group['Price'].sum()
user_avg = user_total / user_purchases

# Create DataFrame
top_spenders_df = pd.DataFrame({
    'Purchase Count':user_purchases,
    'Average Purchase Price':user_avg,
    'Total Purchase Value':user_total},
    index = binned_df['SN'])

# Format DataFrame
top_spenders_df = top_spenders_df.sort_values('Total Purchase Value', ascending=False)
top_spenders_df['Average Purchase Price'] = top_spenders_df['Average Purchase Price'].map("${:.2f}".format)
top_spenders_df['Total Purchase Value'] = top_spenders_df['Total Purchase Value'].map("${:.2f}".format)

# Show summary
top_spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Make DataFrame with necessary columns
items_df = purchase_df[['Item ID', 'Item Name', 'Price']]

# Group by Item ID and by Name
item_groups = items_df.groupby(['Item ID', 'Item Name'])

# Variables to hold data for easy use later
pur_count = item_groups['Price'].count()
tot_pur_value = item_groups['Price'].sum()
item_price = item_groups['Price'].mean()

# DataFrame with requirements for Most Popular Items
most_popular_items_df = pd.DataFrame({
    'Purchase Count':pur_count,
    'Item Price':item_price,
    'Total Purchase Value':tot_pur_value
})

# Copy DataFrame for use in next activity
mpi_df = most_popular_items_df

# Format DataFrame
most_popular_items_df = most_popular_items_df.sort_values('Purchase Count', ascending=False)
most_popular_items_df['Item Price'] = most_popular_items_df['Item Price'].map("${:.2f}".format)
most_popular_items_df['Total Purchase Value'] = most_popular_items_df['Total Purchase Value'].map("${:.2f}".format)

# Show Summary
most_popular_items_df.head()
```
